```python
# read the txt file into a pandas dataframe and append all three files and write into a new txt file
import pandas as pd
# Function to read a file line by line into a DataFrame
def read_file_to_df(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return pd.DataFrame(lines, columns=['text'])

# Read the text files into pandas DataFrames
df1 = read_file_to_df('/home/zichengx/Research/narrativesBERT/output/preprocessed_docs_2010_2013.txt')
df2 = read_file_to_df('/home/zichengx/Research/narrativesBERT/output/preprocessed_docs_2014_2017.txt')
df3 = read_file_to_df('/home/zichengx/Research/narrativesBERT/output/preprocessed_docs_2018_2020.txt')

# Append the DataFrames
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
import csv

# give the column names
combined_df.columns = ['text']
# reduce the duplicates 
combined_df.drop_duplicates(subset=['text'], keep='first', inplace=True)
# Write the combined DataFrame into a new text file
# Write each line of the DataFrame to a separate line in the text file
output_file_path = '/home/zichengx/Research/narrativesBERT/output/preprocessed_docs_2010_2020.txt'
with open(output_file_path, 'w') as file:
    for line in combined_df['text']:
        file.write(line) 
```
